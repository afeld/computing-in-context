## Create a homepage

1. Create a file called `index.md`.
1. Copy and paste the following:

    ```markdown
    # [Name]'s Portfolio Site
    
    Work in progress! Created from the [SIPA Portfolio Site Workshop](https://computing-in-context.afeld.me/portfolio_site_workshop.html).
    ```

You can make this better later.

### Send that to your repository

1. In the side bar, open [Source Control](https://code.visualstudio.com/docs/sourcecontrol/overview).
1. Under `Changes`, you should see `index.md`.
1. Type a meaningful commit message.
1. Click [Commit](https://code.visualstudio.com/docs/sourcecontrol/overview#_commit).
    - If it asks about `There are no staged changes`, click `Always`.
1. Click `Sync Changes`.
    - If it says `This action will pull and push commits from and to "origin/main"`, click `Ok, Don't Show Again`.
1. In another tab, open your repository.

## [Jupyter Book](https://jupyterbook.org/)

- [Static website builder](https://jupyterbook.org/v1/publish/web.html)
- Used to build this site


### Install

1. Create a `requirements.txt`.
1. Add:

    ```
    jupyter-book==2.*
    ```

1. [Open the integrated terminal.](https://code.visualstudio.com/docs/terminal/getting-started)
1. Install the packages.

    ```sh
    pip install -r requirements.txt
    ```


### [Initialize the site](https://jupyterbook.org/stable/get-started/init/)

```sh
jupyter book init --write-toc
```


### [Preview the site](https://jupyterbook.org/stable/get-started/build-websites/)

1. Run:

    ```sh
    jupyter book start
    ```

1. A notification should pop up.
1. Click `Open in Browser`.
    - Make sure you're going to `3000`.
1. [View the diff.](https://code.visualstudio.com/docs/sourcecontrol/overview#_viewing-diffs)
1. Commit.

This will only work for us, while the Codespace is running. We need something public, indefinitely.
